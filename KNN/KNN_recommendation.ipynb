```python
#this code predicts the housing price with a artificial neural network
import pandas as pd
import numpy as np

#read_data from 
data = pd.read_csv("./kc_house_data.csv")

#delete unrelevant features: house id and date
data = data.drop(['id', 'date','lat','long','waterfront','zipcode'],axis=1)
data.head()
```

```python
x_train = data.values
#x_train = (x_train - np.mean(x_train,axis=0))/np.std(x_train,axis=0)

def KNN(data_train,data_test,k=5):
    dist = np.sum(np.power(data_train - data_test,2),axis=1)
    min_index = np.argsort(dist)
    
    return data_train[min_index[0:k]]
#user prefernce
user_preference = [450000,3,2,1500,7000,1.0,0,4,6,1500,400,1980,1,1500,5000]

recommendation = KNN(x_train,user_preference)
print(recommendation)
```
